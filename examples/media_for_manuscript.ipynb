```python
import parafields
import numpy as np

field_ls = []
grid = 128
cov_model = 'gaussian'
min_, max_ = 1e9, 1e-9
for var in {0.1, 0.5, 0.7}:
    coeff = np.exp(parafields.generate_field(cells=[grid+1, grid+1], covariance=cov_model, variance=var).evaluate())
    field_ls.append(coeff)
    min_ = np.min(coeff) if np.min(coeff) < min_ else min_
    max_ = np.max(coeff) if np.max(coeff) > max_ else max_
```

```python
plt.rcParams['font.size'] = 12
plt.rcParams['font.style'] = 'normal'
plt.rcParams['axes.linewidth'] = 1.5#0.8

from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.colors import LogNorm

def make_colorbar(fig, im, ax):
    divider = make_axes_locatable(ax)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    fig.colorbar(im, cax=cax, orientation='vertical')
    return

def plots_3(field_ls, min_, max_, figsize, coords_colorbar):
    fig, axes = plt.subplots(1, 3, figsize=figsize)

    im1 = axes[0].imshow(field_ls[0], vmin=min_, vmax=max_)#norm=LogNorm(vmin=min_, vmax=max_))
    axes[0].set_title('Variance 0.1')

    im2 = axes[1].imshow(field_ls[1], vmin=min_, vmax=max_)#norm=LogNorm(vmin=min_, vmax=max_))
    axes[1].set_title('Variance 0.5')

    im3 = axes[2].imshow(field_ls[2], vmin=min_, vmax=max_)#norm=LogNorm(vmin=min_, vmax=max_))#vmin=min_, vmax=max_)
    axes[2].set_title('Variance 0.7')
            
#     make_colorbar(fig, im1, axes[0], False)
#     make_colorbar(fig, im2, axes[1], False)
#     make_colorbar(fig, im3, axes[2])

    cbar_ax = fig.add_axes(coords_colorbar)
    fig.colorbar(im3, cax=cbar_ax)
    
    axes = axes.flatten()
    for ax in axes:
        ax.set_yticks([], [])
        ax.set_xticks([], [])
        
    plt.tight_layout()
    plt.show()
    return
```

```python
plt.set_cmap('rainbow')
plots_3(field_ls, min_, max_, figsize=(12, 7), coords_colorbar=[1., 0.231, 0.012, 0.537])
```
