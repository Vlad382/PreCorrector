```python
plt.rcParams['font.size'] = 10
_, axes = plt.subplots(len(config_g64), 3, figsize=(20, 20))

for i, c in enumerate(config_g64):
    check = jnp.load(os.path.join(dir_, date, c['run_name']))
    n = c['run_name'][4:c['run_name'].find('.')]
    losses = check['losses']
    R_I = check['res_I']
    R_LLT = check['res_LLT']

    axes[i, 0].plot(range(len(losses[0])), losses[1], label='Test')
    axes[i, 0].plot(range(len(losses[0])), losses[0], label='Train')
    axes[i, 0].legend(fontsize=7)
    axes[i, 0].set_yscale('log')
    axes[i, 0].set_xlabel('Epoch')
    axes[i, 0].set_ylabel(f'{n}\nLoss')
    axes[i, 0].grid()
    
    try:
        axes[i, 1].plot(range(len(losses[0])), losses[2], label='Test')
        axes[i, 1].legend(fontsize=7)
        axes[i, 1].set_yscale('log')
        axes[i, 1].set_xlabel('Epoch')
        axes[i, 1].set_ylabel('Cond of $LL^T$')
        axes[i, 1].grid();
    except: pass
    
    axes[i, 2].plot(range(R_I.shape[-1]), R_I, label="Not prec")
    axes[i, 2].plot(range(R_LLT.shape[-1]), R_LLT, label="$\|LL^Tx - b\|_2^2$")

    axes[i, 2].set_xlabel('Iteration')
    axes[i, 2].set_ylabel('Norm residuals')
    axes[i, 2].legend(fontsize=7);
    axes[i, 2].set_yscale('log')
    axes[i, 2].grid();
plt.tight_layout;
```
